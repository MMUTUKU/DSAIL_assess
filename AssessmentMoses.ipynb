### Question 2

DSAIL is recruiting undergraduate research assistants. Only one-half of those who got shortlisted for the first interview made it to the second stage of the recruitment. Of those invited for the second stage, 90% feel that the first interview went well, whereas only 70% of those **not invited** feel they had a good first interview.

Suppose you feel that your first interview went well:

---

#### (a) Calculate the probability that you will get invited to the second stage.

Given:

- ( P(I) = $\frac{1}{2}$ \) (Probability of being invited to the second stage)
- ( P(W \I) = 0.90 ) (Probability of feeling good about the interview, given you were invited)
- ( P(W \-I) = 0.70 ) (Probability of feeling good about the interview, given you were **not** invited)

You feel the interview went well, so  P(W) .

We need to calculate  P(I \ W) , the probability of being invited given that you feel good about the interview. Using **Bayes' Theorem**:

$
P(I\W) = \frac{P(W \mid I) \cdot P(I)}{P(W)}
$

Where  P(W) is the total probability of feeling good about the interview:

$
P(W) = P(W \mid I) \cdot P(I) + P(W \mid \neg I) \cdot P(\neg I)
$

Now, substituting the values:

$
P(W) = (0.90 \cdot \frac{1}{2}) + (0.70 \cdot \frac{1}{2}) = 0.80
$

Then:

$
P(I \mid W) = \frac{0.90 \cdot \frac{1}{2}}{0.80} = 0.5625
$

---

So, the probability of being invited to the second stage, given that you feel good about the interview, is **0.5625** (or **56.25%**).


#### (b) Does it help to feel good about the interview?
Based on the result from part (a), it does **slightly** increase the probability of being invited (56.25% vs 50%). *However, the feeling is not a strong indicator since the probability is not much higher than random chance.*


#### (3) Write a Python script that prompts the user to enter a bird species, then downloads and saves audio files of that species from Xeno-Canto. The program should select files that contain thesp ecified species only

````python
import requests
import os

def download_bird_audio(species):
    url = f"https://www.xeno-canto.org/api/2/recordings?query={species}"
    response = requests.get(url)
    data = response.json()
    
    if 'recordings' in data:
        for index, recording in enumerate(data['recordings']):
            if species.lower() in recording['en'].lower():
                audio_url = recording['file']
                audio_data = requests.get(audio_url)
                
                filename = f"{species}_audio_{index+1}.mp3"
                with open(filename, 'wb') as file:
                    file.write(audio_data.content)
                print(f"Downloaded: {filename}")
    else:
        print(f"No recordings found for species: {species}")

if __name__ == "__main__":
    bird_species = input("Enter bird species: ")
    download_bird_audio(bird_species)
````

#### (a) Derive the sigmoid function.

The **odds** are defined as the ratio of the probability that an event occurs to the probability that it doesn't occur:

$
\text{odds} = \frac{p}{1 - p}
$

Taking the natural logarithm of both sides gives the **log-odds**:

$
\log\left(\frac{p}{1 - p}\right)
$

We assume that the log-odds follow a linear relationship with the predictor variables, represented as $( z )$. Hence,

$
\log\left(\frac{p}{1 - p}\right) = z
$

Solving for $( p )$:

$
p = \frac{1}{1 + e^{-z}}
$

This is the **sigmoid function**, where $( z )$ is the linear combination of predictor variables, and $( p)$ is the probability output between 0 and 1.

#### (b) Plot the sigmoid curve for a range of values (-15 to 15).

The Python code below can be used to generate a plot of the sigmoid function over the range of \( z \) values from -15 to 15:

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Generate values from -15 to 15
z = np.linspace(-15, 15, 300)
p = sigmoid(z)

# Plot the sigmoid curve
plt.plot(z, p)
plt.title("Sigmoid Function")

plt.ylabel("Sigmoid(z)")

````python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Generate values from -15 to 15
z = np.linspace(-15, 15, 300)
p = sigmoid(z)

# Plot the sigmoid curve
plt.plot(z, p)
plt.title("Sigmoid Function")
plt.xlabel("z")
plt.ylabel("Sigmoid(z)")
plt.grid(True)
plt.show()
````
